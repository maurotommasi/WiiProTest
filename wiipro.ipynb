```python
# Create a DB
wp.inizialize_sqlite3() # mokeup data disabled by default
```

```python
# This is a custom function that can be applied to the INSTRUMENT3 data
def std(column_df):
    return column_df.std()
```

```python
# The engine can be run one time at day, being the analysis made in a daily basis concept
# This engine is optimized in a multithread environment
report = wp.run_engine(std) 
```

```python
# Show the result of the elaboration
print(report)
```
